```python
# load iris data
import pandas as pd
import numpy as np
data = pd.read_csv('iris_data.csv')
data.head()
```

```python
# define X and y
X = data.drop(['target', 'label'],axis = 1)
y = data.loc[:, 'label']
print(X.shape, y.shape)
```

```python
# establish the decision tree model
# min samples leaf is 10
from sklearn import tree
dc_tree = tree.DecisionTreeClassifier(criterion='entropy', min_samples_leaf=5)
dc_tree.fit(X, y)
```

```python
# evaluate the model
from sklearn.metrics import accuracy_score

y_predict = dc_tree.predict(X)
print(accuracy_score(y, y_predict))
```

```python
# visualize the decision tree
%matplotlib inline
from matplotlib import pyplot as plt
fig = plt.figure(figsize=(20,20))
tree.plot_tree(dc_tree,filled=True,feature_names=['SepalLength', 'SepalWidth', 'PetalLength', 'PetalWidth'],class_names=['setosa','versicolor','virginica'])
```

```python
fig.savefig('dc_tree_leaf_5.png')
```

```python
from IPython import display
display.Image(filename="./dc_tree_leaf_5.png", width=800)
```

```python
# min samples leaf is 10
dc_tree = tree.DecisionTreeClassifier(criterion='entropy',min_samples_leaf=10)
dc_tree.fit(X,y)
```

```python
y_predict_10 = dc_tree.predict(X)
print(accuracy_score(y, y_predict_10))
```
